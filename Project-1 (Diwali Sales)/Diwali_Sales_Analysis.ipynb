# Exploratory Data Analysis

### Gender

```python
# plotting a bar chart for Gender and it's count

gender_counts = df['Gender'].value_counts()

gender_labels = gender_counts.index
gender_values = gender_counts.values

plt.figure(figsize=(8, 6))
bars = plt.bar(gender_labels, gender_values, color=['pink', 'blue', 'green'])

plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Count of Each Gender')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 0.2, 
             f'{height}', 
             ha='center', va='bottom')

plt.show()
```

```python
# plotting a bar chart for gender vs total amount

gender_amounts = df.groupby('Gender')['Amount'].sum()

gender_labels = gender_amounts.index
gender_values = gender_amounts.values

plt.figure(figsize=(8, 6))
bars = plt.bar(gender_labels, gender_values, color=['pink', 'blue', 'green'])

plt.xlabel('Gender')
plt.ylabel('Total Amount')
plt.title('Total Amount by Gender')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 10,  
             f'{height}', 
             ha='center', va='bottom')

plt.show()
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

### Age

```python
# plotting a bar chart for count of age groups based on gender

bins = [0, 18, 30, 40, 50, 60, 100]  
labels = ['0-18', '19-30', '31-40', '41-50', '51-60', '61+']

df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

age_gender_counts = df.groupby(['AgeGroup', 'Gender']).size().unstack(fill_value=0)
age_gender_counts.plot(kind='bar', figsize=(10, 6), colormap='viridis')

plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Count of Age Groups by Gender')

for i in range(len(age_gender_counts.index)):
    for j, value in enumerate(age_gender_counts.iloc[i]):
        plt.text(i, value , f'{value}', ha='center', va='bottom')

plt.show()
```

```python
# plotting a bar chart for total amount based on age groups

age_group_totals = df.groupby('AgeGroup')['Amount'].sum()

age_groups = age_group_totals.index
total_amounts = age_group_totals.values

plt.figure(figsize=(10, 6))
bars = plt.bar(age_groups,total_amounts, color='skyblue')

plt.xlabel('Age Group')
plt.ylabel('Total Amount')
plt.title('Total Amount by Age Group')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, 
             f'{height}', 
             ha='center', va='bottom')

plt.show()
```

*From above graphs we can see that most of the buyers are of age group between 31-40 yrs female*

### State

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

plt.figure(figsize=(12, 8))  
bars = plt.bar(sales_state['State'], sales_state['Orders'], color='skyblue')

plt.title('Total Number of Orders from Top 10 States')
plt.xlabel('States')
plt.ylabel('Total Number of Orders')

plt.xticks(rotation=45)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, int(yval),ha='center', va='bottom') 

plt.tight_layout()  
plt.show()
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

plt.figure(figsize=(12, 8))  
bars = plt.bar(sales_state['State'], sales_state['Amount'], color='blue')

plt.title('Total sales from Top 10 States')
plt.xlabel('States')
plt.ylabel('amount')

plt.xticks(rotation=45)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, int(yval),ha='center', va='bottom') 

plt.tight_layout()  
plt.show()
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*


### Marital Status

```python
# total count by marital status

marital_status_counts = df['Marital_Status'].value_counts()

plt.figure(figsize=(8, 8))

new_labels = ['Married', 'Unmarried']

plt.pie(marital_status_counts,labels=new_labels,autopct='%1.1f%%',colors=['skyblue','salmon'])

plt.title('Distribution by Marital Status')

plt.tight_layout()
plt.show()
```

*From above graphs we can see that most of the buyers are married and they have high purchasing power*

### Occupation

```python
# total count by Occupations

occ_counts = df['Occupation'].value_counts()

occ_labels = occ_counts.index
occ_values = occ_counts.values

plt.figure(figsize=(20, 6))
bars = plt.bar(occ_labels, occ_values)

plt.xticks(rotation=90)

plt.xlabel('Occupations')
plt.ylabel('Count')
plt.title('Count of Each Occupation')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 0.2, 
             f'{height}', 
             ha='center', va='bottom')

plt.show()
```

```python
# total amount by Occupations

occ_amounts = df.groupby('Occupation')['Amount'].sum()

occ_labels = occ_amounts.index
occ_values = occ_amounts.values

plt.figure(figsize=(20, 6))
bars = plt.bar(occ_labels,occ_values)

plt.xticks(rotation=90)

plt.xlabel('Occupation')
plt.ylabel('Total Amount')
plt.title('Total Amount by Occupation')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 10,  
             f'{height}', 
             ha='center', va='bottom')

plt.show()
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

### Product Category

```python
# Most sold product catagories

product_counts = df['Product_Category'].value_counts()

product_labels = product_counts.index
product_values = product_counts.values

plt.figure(figsize=(15, 6))
bars = plt.bar(product_labels,product_values)

plt.xticks(rotation=90)

plt.xlabel('Product Category')
plt.ylabel('Count')
plt.title('Count of Each Category')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 0.2, 
             f'{height}', 
             ha='center', va='bottom')

plt.show()
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*
