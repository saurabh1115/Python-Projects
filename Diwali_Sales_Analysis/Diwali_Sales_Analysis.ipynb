# Exploratory Data Analysis

### Gender

```python
# plotting a bar chart for Gender and it's count

# Get the counts of each gender
gender_counts = df['Gender'].value_counts()

# Extract the labels and values
gender_labels = gender_counts.index
gender_values = gender_counts.values

# Create the bar plot
plt.figure(figsize=(8, 6))

# Adjust the number of colors dynamically based on the number of gender categories
colors = ['pink', 'blue', 'green', 'orange', 'purple'][:len(gender_labels)]
bars = plt.bar(gender_labels, gender_values, color=colors)

# Add labels and title
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Count of Each Gender')

# Adding text labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 0.2, 
             f'{int(height)}',  # Ensuring the value is displayed as an integer
             ha='center', va='bottom')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# plotting a bar chart for gender vs total amount

# Grouping by 'Gender' and summing up the 'Amount'
gender_amounts = df.groupby('Gender')['Amount'].sum()

# Extract labels and values
gender_labels = gender_amounts.index
gender_values = gender_amounts.values

# Create the bar plot
plt.figure(figsize=(8, 6))

# Dynamic color adjustment based on number of gender categories
colors = ['pink', 'blue', 'green', 'orange', 'purple'][:len(gender_labels)]
bars = plt.bar(gender_labels, gender_values, color=colors)

# Add labels and title
plt.xlabel('Gender')
plt.ylabel('Total Amount')
plt.title('Total Amount by Gender')

# Adding text labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + (0.02 * height),  # Offset adjusted dynamically
             f'{height:,.2f}',  # Formatting with commas and 2 decimal places
             ha='center', va='bottom')

# Show the plot
plt.tight_layout()
plt.show()
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

### Age

```python
# plotting a bar chart for count of age groups based on gender

# Define age bins and labels
bins = [0, 18, 30, 40, 50, 60, 100]  
labels = ['0-18', '19-30', '31-40', '41-50', '51-60', '61+']

# Add 'AgeGroup' column to the DataFrame
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Group by 'AgeGroup' and 'Gender', then unstack the gender counts with observed=True
age_gender_counts = df.groupby(['AgeGroup', 'Gender'], observed=True).size().unstack(fill_value=0)

# Create the bar plot
ax = age_gender_counts.plot(kind='bar', figsize=(10, 6), colormap=plt.cm.viridis)

# Add labels and title
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Count of Age Groups by Gender')

# Adding text labels on top of each bar
for i in range(len(age_gender_counts.index)):
    for j, value in enumerate(age_gender_counts.iloc[i]):
        ax.text(i + j * 0.15, value + 1,  # Offset slightly to avoid overlapping
                f'{value}', ha='center', va='bottom', fontsize=10)

# Adjust layout to avoid clipping
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# plotting a bar chart for total amount based on age groups

# Grouping data by 'AgeGroup' and summing up 'Amount'
age_group_totals = df.groupby('AgeGroup', observed=True)['Amount'].sum()

# Extracting age groups and total amounts
age_groups = age_group_totals.index
total_amounts = age_group_totals.values

# Creating the bar plot
plt.figure(figsize=(10, 6))
bars = plt.bar(age_groups, total_amounts, color='skyblue')

# Adding labels and title
plt.xlabel('Age Group')
plt.ylabel('Total Amount')
plt.title('Total Amount by Age Group')

# Adding text labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + (0.02 * height),  # Slight offset to avoid overlap
             f'{height:,.2f}',  # Formatting with commas and 2 decimal places
             ha='center', va='bottom')

# Display the plot
plt.tight_layout()
plt.show()
```

*From above graphs we can see that most of the buyers are of age group between 31-40 yrs female*

### State

```python
# total number of orders from top 10 states

# Grouping data by 'State', summing 'Orders', and sorting by 'Orders'
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# Creating the bar plot
plt.figure(figsize=(12, 8))  
bars = plt.bar(sales_state['State'], sales_state['Orders'], color='skyblue')

# Adding title and labels
plt.title('Total Number of Orders from Top 10 States')
plt.xlabel('States')
plt.ylabel('Total Number of Orders')

# Rotate the x-axis labels for readability
plt.xticks(rotation=45, ha='right')

# Adding text labels on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + (0.02 * yval),  # Small offset for clarity
             f'{int(yval):,}',  # Formatting the number with commas
             ha='center', va='bottom')

# Ensure everything fits well
plt.tight_layout()  

# Show the plot
plt.show()
```

```python
# total amount/sales from top 10 states

# Grouping by 'State', summing 'Amount', and sorting by 'Amount' to get top 10 states
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

# Create the bar plot
plt.figure(figsize=(12, 8))  
bars = plt.bar(sales_state['State'], sales_state['Amount'], color='blue')

# Adding title and labels
plt.title('Total Sales from Top 10 States')
plt.xlabel('States')
plt.ylabel('Total Amount')

# Rotate the x-axis labels for readability
plt.xticks(rotation=45, ha='right')

# Adding text labels on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + (0.02 * yval),  # Adding a slight offset
             f'{int(yval):,}',  # Formatting the number with commas for readability
             ha='center', va='bottom')

# Ensure everything fits well
plt.tight_layout()  

# Show the plot
plt.show()
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*


### Marital Status

```python
# total count by marital status

# Getting the counts of each marital status
marital_status_counts = df['Marital_Status'].value_counts()

# Dynamically extracting the labels from the DataFrame
labels = ["Married","Unmarried"]

# Creating the pie chart
plt.figure(figsize=(5, 5))  # Slightly adjusted the size for better visibility

plt.pie(marital_status_counts, labels=labels, autopct='%1.1f%%', 
        colors=['skyblue', 'salmon'], startangle=90, wedgeprops={'edgecolor': 'black'})

# Adding a title
plt.title('Distribution by Marital Status')

# Ensuring the layout is well-fitted
plt.tight_layout()

# Show the plot
plt.show()
```
